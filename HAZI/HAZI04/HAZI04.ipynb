```python
def capitalize_columns(df:pd.core.frame.DataFrame)->pd.core.frame.DataFrame:
    ndf = df.copy()
    ndf = ndf.rename(columns=renaming_str)
    return ndf

def renaming_str(x:str):
    if "e" in x or "E" in x:
        return x.capitalize()
    return x

#capitalize_columns(df)

```
